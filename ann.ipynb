```python
df=pd.read_csv("new2.csv")
df.head(4)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["CustomerID"],inplace=True) 
```

```python
df.info()
```

```python
x=df.drop("Exited",axis=1)
y=df["Exited"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_new=sc.fit_transform(x_train)
x_new2=sc.transform(x_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(3,activation="relu",input_dim=7))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy",optimizer="Adam")
```

```python
model.fit(x_new,y_train,epochs=10)
```

```python
model.layers[0].get_weights()
```

```python
new3=model.predict(x_new2)
```

```python
import numpy as np
```

```python
y_pred=np.where(new3>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
